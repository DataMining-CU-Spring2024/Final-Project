### Downloading The Datasets

This notebook contains code snippets to extract and download the Amazon Review Data files. 

```python
import requests
from bs4 import BeautifulSoup


# Reading the html file using requests
url = "https://nijianmo.github.io/amazon/index.html"
html_file = requests.get(url)
html_file.content
```

#### Pass the Html Content into Beautiful Soup for Parsing.

0. Create a `urls_for_data_download.txt` data
1. Pass the html content into BeautifulSoup
2. Extract all `<a>` tags
3. From all `<a>` tages, extract `href`


```python
html = BeautifulSoup(html_file.content, "html.parser")
anchor_elements = html.find_all('a')

with open('urls_for_data_download.txt', 'w+') as file:
    for element in anchor_elements:
        if element.get('href') is None:
            pass
        elif element.get('href').startswith('https://jmcauley.ucsd.edu/data/amazon_v2/categoryFilesSmall'):
            file.write(element.get('href'))
            file.write('\n')
```

### Downloading Files

```python
from urllib.request import urlretrieve

with open('urls_for_data_download.txt', 'r+') as download_file:
    files = download_file.readlines()
    for file in files:
        file = file.replace('\n', '')
        basename = file.split('/')[-1]
        urlretrieve(file, basename)
```
